```python
# Two sum problem

# 1. Define the Class (Same as LeetCode)
class Solution:
    def twoSum(self, nums, target):
        prev_map = {} 
        for i, n in enumerate(nums):
            diff = target - n
            if diff in prev_map:
                return [prev_map[diff], i]
            prev_map[n] = i

# 2. Driver Code (The part you run in your cell)
sol = Solution()

# Test Case 1
test_nums = [2, 7, 11, 15]
test_target = 9
result = sol.twoSum(test_nums, test_target)

print(f"Input Array: {test_nums}")
print(f"Target: {test_target}")
print(f"Result Indices: {result}") 
# Expected Output: [0, 1]
```
